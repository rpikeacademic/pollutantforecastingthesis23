```python
stats = {'n':[],'min':[],'max':[],'mean':[],'std':[]}
for n in range(1, ldists.shape[1]):
  lds = ldists[:,:n]
  stats['n'].append(n)
  stats['min'].append(lds.min())
  stats['max'].append(lds.max())
  stats['mean'].append(lds.mean())
  stats['std'].append(lds.std())
ax = plt.gca()
# for k, v in stats.items():
#   if k == 'n': continue
#   ax.plot(stats['n'], v, label=k )#+' distance')
ax.plot(stats['n'], stats['max'], label='max distance', color='gray')
ax.plot(stats['n'], stats['mean'], label='mean distance', color='tab:blue')
y1 = np.array(stats['mean'])-np.array(stats['std'])
y2 = np.array(stats['mean'])+np.array(stats['std'])
ax.fill_between(stats['n'],
                y1,
                y2,
                alpha=0.25, label='std. dev.' )#+' distance')
ax.set_ylabel('Distance (km)')
ax.set_xlabel('$n$ (number of nearest neighbors)')
plt.legend()
plt.title('Distance to $n$ nearest neighbor stations')
ax.set_ylim(0,1600)
ax.set_xlim(0,ldists.shape[1])
```
